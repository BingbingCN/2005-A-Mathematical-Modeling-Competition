```python
#首先计算单因子水质标识指数
'''
溶解氧 7.5 6 5 3 2 0
高锰酸盐指数 2 4 6 10 15 000
氨氮 0.15 0.5 1.0 1.5 2.0
PH 6-9
'''
def data_process(df1_1):
    O2=[]
    Mn=[]
    NH=[]
    PH=[]
    index=df1_1['水环境功能区类别'].values[0]
    for i in df1_1['CODMn'].values:
        if float(i)>=7.5:
            x_1=1
            x_2=int((1-(float(i)-7.5)/10)*10)
            if x_1<=index:
                x_3=0
            else:
                if x_2==0:
                    x_3=3-index-1
                else:
                    x_3=3-index

        elif 6<=float(i)<7.5:
            x_1=2
            x_2=int((1-(float(i)-6)/1.5)*10)
            if x_1<=index:
                x_3=0
            else:
                if x_2==0:
                    x_3=3-index-1
                else:
                    x_3=3-index
        elif 5<=float(i)<6:
            x_1=3
            x_2=int(10*(1-(float(i)-5)/1))
            if x_1<=index:
                x_3=0
            else:
                if x_2==0:
                    x_3=3-index-1
                else:
                    x_3=3-index
        elif 3<=float(i)<5:
            x_1=4
            x_2=int(10*(1-(float(i)-3)/2))
            if x_1<=index:
                x_3=0
            else:
                if x_2==0:
                    x_3=3-index-1
                else:
                    x_3=3-index
        elif 2<=float(i)<3:
            x_1=5
            x_2=int(10*(1-(float(i)-2)/1))
            if x_1<=index:
                x_3=0
            else:
                if x_2==0:
                    x_3=3-index-1
                else:
                    x_3=3-index

        x=str(x_1)+'.'+str(x_2)+str(x_3)
        O2.append(float(x))
    #高锰酸盐  2 4 6 10 15
    for i in df1_1['NH3_N'].values:
        if float(i)<=2:
            x_1=1
            x_2=int((1-(2-float(i))/2.0001)*10)
            if x_1<=index:
                x_3=0
            else:
                if x_2==0:
                    x_3=3-index-1
                else:
                    x_3=3-index

        elif 2<float(i)<=6:
            x_1=2
            x_2=int((1-(6-float(i))/4)*10)
            if x_1<=index:
                x_3=0
            else:
                if x_2==0:
                    x_3=3-index-1
                else:
                    x_3=3-index
        elif 6<float(i)<=10:
            x_1=3
            x_2=int((1-(10-float(i))/6)*10)
            if x_1<=index:
                x_3=0
            else:
                if x_2==0:
                    x_3=3-index-1
                else:
                    x_3=3-index
        elif 10<float(i)<=15:
            x_1=4
            x_2=int((1-(15-float(i))/5)*10)
            if x_1<=index:
                x_3=0
            else:
                if x_2==0:
                    x_3=3-index-1
                else:
                    x_3=3-index
        elif 15<float(i):
            x_1=5
            x_2=int((1-(20-float(i))/5)*10)
            if x_1<=index:
                x_3=0
            else:
                if x_2==0:
                    x_3=3-index-1
                else:
                    x_3=3-index

        x=str(x_1)+'.'+str(x_2)+str(x_3)
        Mn.append(float(x))
    #NH 0.15 0.5 1.0 1.5 2.0 注意导数据时候的问题
    for i in df1_1['PH'].values:
        if float(i)<=0.15:
            x_1=1
            x_2=int((1-(0.15-float(i)+0.0001)/0.1501)*10)
            if x_1<=index:
                x_3=0
            else:
                if x_2==0:
                    x_3=3-index-1
                else:
                    x_3=3-index

        elif 0.15<float(i)<=0.5:
            x_1=2
            x_2=int((1-(0.5-float(i)+0.0001)/0.35)*10)
            if x_1<=index:
                x_3=0
            else:
                if x_2==0:
                    x_3=3-index-1
                else:
                    x_3=3-index
        elif 0.5<float(i)<=1.0:
            x_1=3
            x_2=int((1-(1-float(i)+0.0001)/0.5)*10)
            if x_1<=index:
                x_3=0
            else:
                if x_2==0:
                    x_3=3-index-1
                else:
                    x_3=3-index
        elif 1.0<float(i)<=1.5:
            x_1=4
            x_2=int((1-(1.5-float(i)+0.0001)/0.5)*10)
            if x_1<=index:
                x_3=0
            else:
                if x_2==0:
                    x_3=3-index-1
                else:
                    x_3=3-index
        elif 1.5<float(i)<=2.0:
            x_1=5
            x_2=int((1-(2-float(i)+0.0001)/0.5)*10)
            if x_1<=index:
                x_3=0
            else:
                if x_2==0:
                    x_3=3-index-1
                else:
                    x_3=3-index

        x=str(x_1)+'.'+str(x_2)+str(x_3)
        NH.append(float(x))
    return O2,Mn,NH
```
